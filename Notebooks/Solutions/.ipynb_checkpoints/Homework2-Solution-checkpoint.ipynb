## Problem 3. Learning of experimental bandgaps

We will use a dataset from Materials Project that includes over 4,000 experimental bandgaps of materials. Your task is to optimize several machine learning models the dataset and validate the prediction error.

a) 

b) Generate a random 80/20 train-test split and remove the target property from the training and testing data.

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
from sklearn import decomposition
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV

from matminer.datasets import load_dataset
from matminer.featurizers.base import MultipleFeaturizer
from matminer.featurizers import composition as cf
from matminer.featurizers.conversions import StrToComposition

import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15)

data = load_dataset("expt_gap")
data.describe()
```

```python
data.head()
```

```python
# How many unique compositions do we have?
data["formula"].unique().shape[0]
```

```python
# Sort by size of badgap
data = data.sort_values('gap expt')

# Remove duplicate compositions
data = data.drop_duplicates('formula')

data.describe()
```

### Obtain a Feature Vector for Each Material

The first step in building a machine learning model is to convert the raw materials data, here the composition, into the required input for an ML model: a finite list of quantitative attributes. Here we use the Magpie descriptors from Ward et al.

```python
feature_calculators = MultipleFeaturizer([cf.Stoichiometry(), cf.ElementProperty.from_preset("magpie"),
                                          cf.ValenceOrbital(props=['avg']), cf.IonProperty(fast=True)])

# Get the feature names
feature_labels = feature_calculators.feature_labels()

# Compute the features for all materials entries
data = StrToComposition(target_col_id='composition_obj').featurize_dataframe(data, 'formula')
data = feature_calculators.featurize_dataframe(data, col_id='composition_obj', ignore_errors=True);
```

```python
print('Generated %d features'%len(feature_labels))
print('Training set size:', 'x'.join([str(x) for x in data[feature_labels].shape]))
print('Feature labels', feature_labels)
```

```python
# Retain only numerical values
data_num = data.select_dtypes([np.number])
data_num.head(10)
data_num.describe()
```

```python
# Drop the columns that include incomplete data
data_num = data_num.dropna(axis=0)

data_num.describe()
```

### Generate a train-test split

For machine learning, we always split the data into a training and testing set

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data_num, test_size=0.2, shuffle=True, random_state=20220316)
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, cross_val_predict, GridSearchCV, ShuffleSplit, KFold

model = RandomForestRegressor(n_estimators=10, max_features = 5)
```

```python
# Fitting ML model to the dataset
X_train = train_data.loc[:, train_data.columns != 'gap expt']
Y_train = train_data['gap expt']
model.fit(X_train, Y_train)
```

```python
# Regressor score is the coefficient of determination of the prediction
print('Training score =', np.round(model.score(X_train, Y_train),3))
X_test = test_data.loc[:, test_data.columns != 'gap expt']
Y_test = test_data['gap expt']
print('Testing  score =', np.round(model.score(X_test, Y_test),3))
```

```python
from sklearn.metrics import mean_squared_error

Y_train_pred = model.predict(X_train)
training_mse = mean_squared_error(Y_train, Y_train_pred)
X_test_pred = model.predict(X_test)
testing_mse = mean_squared_error(Y_test, X_test_pred)
    
print('Training RMSE = ', np.round(np.sqrt(training_mse),3))
print('Testing  RMSE = ', np.round(np.sqrt(testing_mse),3))
```

```python
# Grid search cross-validation
# Hyper parameters range intialization for tuning
parameters={"min_impurity_decrease" : [0, 0.1, 0.2],
            "min_samples_split" : [2, 3, 4],
            "max_depth" : [None, 4, 6],
            "min_samples_leaf":[1, 2],
            "max_leaf_nodes":[None, 2, 4]}
```

```python
grid_search = GridSearchCV(model,param_grid=parameters,
                            scoring='neg_mean_squared_error',cv=3,verbose=1)
grid_search.fit(X_train, Y_train)
```

```python
# Optimial hyperparameters 
tuned_parameters = grid_search.best_params_
print(tuned_parameters)
```

```python
tuned_regressor = RandomForestRegressor(**tuned_parameters)
tuned_regressor.fit(X_train, Y_train)

print('Training score =', np.round(tuned_regressor.score(X_train,Y_train),3))
print('Testing  score =', np.round(tuned_regressor.score(X_test,Y_test),3))

Y_train_pred = tuned_regressor.predict(X_train)
training_mse = mean_squared_error(Y_train, Y_train_pred)
Y_test_pred = tuned_regressor.predict(X_test)
testing_mse = mean_squared_error(Y_test, Y_test_pred)
    
print('Training RMSE = ', np.round(np.sqrt(training_mse),3))
print('Testing  RMSE = ', np.round(np.sqrt(testing_mse),3))
```
